```python
import pandas as pd
import sqlalchemy as db
```

```python
engine = db.create_engine("postgresql://root:root@localhost/ny_taxi")
```

```python
engine.connect()
```

```python
df_green_taxi = pd.read_csv("green_tripdata_2019-09.csv", nrows=100)
```

```python
df_green_taxi.info(verbose=True)
```

```python
df_green_taxi
```

```python
df_green_taxi.head(0).to_sql(name='green_taxi_data', con=engine, if_exists='replace')
```

```python
df_iters = pd.read_csv('green_tripdata_2019-09.csv', iterator=True, chunksize=50000)
```

```python
next(df_iters)
```

```python
try:
    df_iter = pd.read_csv('green_tripdata_2019-09.csv', iterator=True, chunksize=50000)
    while True:
        df=next(df_iter)
        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
        df.to_sql(name='green_taxi_data', con=engine, if_exists='append')
except StopIteration:
    print("End of Iteration!!")
```
